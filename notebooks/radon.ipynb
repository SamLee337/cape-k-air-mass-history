```python
# All libraries required for analysis
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import string 
import regex
```

```python
rad_df = pd.read_csv('radon.csv')
rad_df['radonmBq'] = rad_df['radon']*1000
rad_df[['date','hour']] = (rad_df['interval_end_utc'].str.split(' ', expand=True))
```

```python
rad_df
```

```python
plt.plot(rad_df.datetime, rad_df.radon)
#More plotting options
#plt.plot(rad_df.datetime, rad_df.radon, marker='o', linestyle='', markersize=8, color='r', label='line Plot')
plt.title('Radon measurement in bq')
plt.xlabel('x')
plt.ylabel('y')
plt.show();
```

```python
rad_df['datetime'] = pd.to_datetime(rad_df['datetime'])
```

```python
df_sept = rad_df[rad_df['datetime'] >'2024-09-17']
df_sept = df_sept[df_sept['datetime'] <'2024-09-21']
```

```python
plt.figure(figsize=(15,8))
plt.plot(df_sept.datetime, df_sept.radonmBq)
#More plotting options
#plt.plot(rad_df.datetime, rad_df.radon, marker='o', linestyle='', markersize=8, color='r', label='line Plot')
plt.title('Radon measurement in mbq')
plt.xlabel('timeperiod surrounding 19 Sept 2024')
plt.ylabel('mBq')
plt.savefig('timeperiod surrounding 19 Sept 2024') 
plt.show();
```
